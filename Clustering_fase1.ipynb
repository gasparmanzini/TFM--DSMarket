Ahora procedemos a unir los items para poder visualizar 

```python
# Calcular el precio promedio y la variabilidad (desviación estándar) por producto
precio_stats = sales_wprice.groupby("item").agg(
    precio_promedio=("unit_price", "mean"),
    variabilidad_precio=("unit_price", "std"),
    total_ventas=("total_venta", "sum"),
    ventas_diarias_x=("total_venta", "mean")
).reset_index()

# Incorporar datos estacionales (e.g., ventas por mes o trimestre)
ventas_estacionales = sales_wprice.groupby(["item", "weekday"]).agg(
    ventas_diarias=("total_venta", "sum")
).reset_index()

# Unificar los datos en un DataFrame
df_productos = precio_stats.merge(ventas_estacionales, on="item", how="left")
df_productos
```
